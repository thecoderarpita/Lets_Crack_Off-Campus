```python
#Leaders in an array
def leaders(arr):
  res = []
  n = len(arr)
  for i in range(n):
    for j in range(i+1,n):
      if arr[i]<arr[j]:
        break
    else:
      res.append(arr[i])
  return res
arr = [16,17,4,3,5,2]
print(leaders(arr))
```
